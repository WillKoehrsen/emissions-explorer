```python
import pandas as pd
import numpy as np
import json
```

```python
df_socio = pd.DataFrame.from_csv('socio_econ.csv', index_col=None)
```

```python
df_socio.fillna(0, inplace=True)
```

```python
with open('country_listing_15.json') as f:
    top_emissions = json.load(f)
```

```python
top_countries = []
for country in top_emissions:
    top_countries.append(country['country'])
```

```python
# Change names to reflect those already found
for i, entry in enumerate(df_socio['country']):
    if entry == 'United States of America':
        df_socio.ix[i, 'country'] = 'United States'
    elif entry == 'Korea (South)':
        df_socio.ix[i, 'country'] = 'South Korea'
    elif entry == 'Russian Federation':
        df_socio.ix[i, 'country'] = 'Russia'
```

```python
# Find indices of top countries in all socioeconomic data
top_indices = [i for i, country in enumerate(list(df_socio['country'])) if country in top_countries]
```

```python
df_top_socio = df_socio.iloc[top_indices, :]
```

```python
top_emitters = pd.DataFrame.from_dict(top_emissions, orient = 'columns')
top_emitters.head(16)
```

```python
with open('emissions_yearly_15.json', 'r') as f:
    emissions_data = json.load(f)
```

```python
codes = []
countries = []
emissions = []
years = []
for entry in emissions_data:
    codes.append(entry['code'])
    countries.append(entry['country'])
    emissions.append(entry['emissions'])
    years.append(entry['year'])
```

```python
emissions_df = pd.DataFrame({'code': codes, 'country': countries,
                             'emission': emissions, 'year': years})
```

```python
df_top_socio = df_top_socio.merge(emissions_df, how = 'left', on = ['country', 'year'])
```

```python
top_socio_dict = df_top_socio.to_dict(orient = 'records')
```

```python
with open("socio_econ_15.json", "w") as out:
    json.dump(top_socio_dict, out)
```
