```python
pop = pop.transpose()
pop.columns = pop.loc['Population']
pop.drop('Population', axis=0, inplace=True)
pop = pop.reset_index(level=0)
pop.rename(columns={'index': 'Year'}, inplace=True)
pop = pd.melt(pop, id_vars='Year', value_name='Pop')
pop.rename(columns={'Population': 'Country'}, inplace=True)
pop.head()
```

```python
gdp['Year'] = gdp['Year'].astype('int64')
gdp.dtypes
```

```python
pop['Year'] = pop['Year'].astype('int64')
pop['Pop'] = pop['Pop'].astype('float64')
pop.dtypes
```

```python
results = pd.merge(co2, gdp, on = ['Year', 'Country'])
results = pd.merge(results, pop, on = ['Year', 'Country'])


results.fillna(0, inplace=True)
results.head()
```

```python
print(results.dtypes)
print(len(results))
```

```python
country = 'Russia'
subset_df = results[results['Country'] == country]
subset_df.ix[(subset_df['Pop'] == 0), 'Pop'] = np.nan
subset_df.ix[subset_df.index[0], 'Pop'] = 0
    
```

```python
subset_df.head()
```

```python
new_results = pd.DataFrame(columns=results.columns)

for country in countries:
    subset_df = results[results['Country'] == country]
    subset_df.ix[(subset_df['Pop'] == 0), 'Pop'] = np.nan
    subset_df.ix[(subset_df['gdp'] == 0), 'gdp'] = np.nan
    subset_df.ix[subset_df.index[0], 'Pop'] = 0
    subset_df.ix[subset_df.index[0], 'gdp'] = 0
    subset_df['Pop'] = subset_df['Pop'].interpolate()
    subset_df['gdp'] = subset_df['gdp'].interpolate()
    new_results = new_results.append(subset_df)
    
new_results.shape
```

```python
new_results['rank'] = new_results.groupby(['Year'])['emissions'].rank(method = 'dense', ascending=False)
```

```python
new_results[(new_results['Year'] > 1980) & (new_results['Pop'] == 0)] 
```

```python
new_results[(new_results['Year'] > 1950) & (new_results['gdp'] == 0)] 
```

```python
new_results[(new_results['Year'] > 1950) & (new_results['Pop'] == 0)] 
```

```python
new_results = new_results.drop_duplicates(subset=['Country', 'Year'])
```

```python
new_results.tail(10)
```

```python
new_results.ix[new_results['rank'] < 4, 'color'] = 'red'
new_results.ix[(new_results['rank'] >= 4) & (new_results['rank'] < 7), 'color'] = 'yellow'
new_results.ix[(new_results['rank'] >= 7) & (new_results['rank'] < 10), 'color'] = 'orange'
new_results.ix[(new_results['rank'] >= 10) & (new_results['rank'] < 13), 'color'] = 'blue'
new_results.ix[new_results['rank'] >= 13, 'color'] = 'green'
```

```python
new_results.rename(columns={'Country': 'country', 'Year': 'year', 'Pop': 'pop'},
                   inplace=True)
new_results.head()
```

```python
new_results = new_results.sort_values(['country', 'year'])
```

```python
len(set(new_results['country']))
```

```python
new_results.shape
```

```python
new_results.head()
```

```python
country_list = []

for country in countries:
    print(country)
    subset_df = new_results[new_results['country'] == country]
    rank_list = []
    gdp_list = []
    color_list = []
    pop_list = []
    emissions_list = []
    for year in subset_df['year']:
        
        rank_list.append([year, float(subset_df.ix[subset_df['year'] == year, 'rank'])])
        gdp_list.append([year, float(subset_df.ix[subset_df['year'] == year, 'gdp'])])
        color = subset_df.ix[subset_df[subset_df['year'] == year].index, 'color']
        color = color.values[0]
        color_list.append([year, color])
        pop_list.append([year, float(subset_df.ix[subset_df['year'] == year, 'pop'])])
        emissions_list.append([year, float(subset_df.ix[subset_df['year'] == year, 'emissions'])])
    
    country_list.append({'name': country, 'rank': rank_list, 
                             'pop': pop_list, 'gdp': gdp_list,
                             'emissions': emissions_list, 'color': color_list})
```

```python
import json

with open('socio_econ.json', 'w') as fp:
    json.dump(country_list, fp)
```
